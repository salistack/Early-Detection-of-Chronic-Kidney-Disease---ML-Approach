```python
df_data.columns = [
    'age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 
    'red_blood_cells', 'pus_cell', 'pus_cell_clumps', 'bacteria', 
    'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium', 
    'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 
    'red_blood_cell_count', 'hypertension', 'diabetes_mellitus', 
    'coronary_artery_disease', 'appetite', 'peda_edema', 'anemia', 'class'
]
```

```python
text_colum =['packed_cell_volume','white_blood_cell_count','red_blood_cell_count']
for i in text_colum:
    print(f"{i} :{df_data[i].dtype}")


```

```python
def convert_text_to_numeric(df_data,colum):
    df_data[colum] = pd.to_numeric(df_data[colum],errors='coerce')

for colum in text_colum:
    convert_text_to_numeric(df_data,colum)
    print(f"{colum} : {df_data[colum].dtype}" )    
```

```python
missing = df_data.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
```

```python
def mean_value_imputation(df_data,colum):
    mean_value = df_data[colum].mean()
    df_data[colum].fillna(value =mean_value)

def mode_value_imputation(df_data,colum):
    mode_value = df_data[colum].mode()[0]
    df_data[colum]=df_data[colum].fillna(mode_value)
        
```

```python
num_cols = [col for col in df_data.columns if df_data[col].dtype != 'object']
for col_num in num_cols:
    mean_value_imputation(df_data,col_num)
```

```python
cat_cols = [col for col in df_data.columns if df_data[col].dtype == 'object']
for col_cat in cat_cols:
    mode_value_imputation(df_data,col_cat)
```

```python
df_data['diabetes_mellitus'] = df_data['diabetes_mellitus'].replace(to_replace={" yes":"yes","\tyes":"yes","\tno":"no"})
df_data['coronary_artery_disease'] = df_data['coronary_artery_disease'].replace(to_replace={"\tno":"no"})
df_data['class'] = df_data['class'].replace(to_replace={"ckd\t":"ckd","notckd":"not ckd"})
```

```python
df_data['class'] = df_data['class'].map({'ckd': 1, 'not ckd': 0})
df_data['red_blood_cells'] = df_data['red_blood_cells'].map({'normal': 1, 'abnormal': 0})
df_data['pus_cell'] = df_data['pus_cell'].map({'normal': 1, 'abnormal': 0})
df_data['pus_cell_clumps'] = df_data['pus_cell_clumps'].map({'present': 1, 'notpresent': 0})
df_data['bacteria'] = df_data['bacteria'].map({'present': 1, 'notpresent': 0})
df_data['hypertension'] = df_data['hypertension'].map({'yes': 1, 'no': 0})
df_data['diabetes_mellitus'] = df_data['diabetes_mellitus'].map({'yes': 1, 'no': 0})
df_data['coronary_artery_disease'] = df_data['coronary_artery_disease'].map({'yes': 1, 'no': 0})
df_data['appetite'] = df_data['appetite'].map({'good': 1, 'poor': 0})
df_data['peda_edema'] = df_data['peda_edema'].map({'yes': 1, 'no': 0})
df_data['anemia'] = df_data['anemia'].map({'yes': 1, 'no': 0})
```

```python
from sklearn.model_selection import train_test_split
```

```python
x= df_data.drop("class", axis=1)
y = df_data["class"]

x_train,x_test,y_train,y_test  = train_test_split(x,y,train_size=0.25,random_state=25)

print(f"'x_train' shape : {x_train.shape}")
print(f"'x_test' shape : {x_test.shape}")
print(f"'y_train' shape : {y_train.shape}")
print(f"'y_test' shape : {y_test.shape}")
```

```python
from sklearn.tree import DecisionTreeClassifier

dct = DecisionTreeClassifier()
dct.fit(x_train,y_train)
```

```python
y_pres_dct= dct.predict(x_test)
```

```python
y_pres_dct
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
models =[]
models.append(('Native bsysis',GaussianNB()))
models.append(('KNN ',KNeighborsClassifier(n_neighbors=8)))
models.append(('RandomForestClassifier',DecisionTreeClassifier()))
models.append(('SVM',SVC(kernel='linear')))
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
for name,model in models:
    print(name,model)
    print()

    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)

    print(confusion_matrix(y_test,y_pred))
    print('\n')

    print("accuracy" ,accuracy_score(y_test,y_pred))
    print('\n')

    print("precision" ,precision_score(y_test,y_pred))
    print('\n')

    print("recall_score",recall_score(y_test,y_pred))
    print('\n')

    print("f1_score" ,f1_score(y_test,y_pred))
    print('\n')

```
